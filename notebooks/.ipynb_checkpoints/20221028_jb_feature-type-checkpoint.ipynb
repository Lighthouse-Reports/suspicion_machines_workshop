```python
import pandas as pd
import numpy as np
```

```python
source_file_ranges = '../data/01_raw/ranges_features_dataset.csv'
source_file_features = '../data/01_raw/20221022_ted_cleaned_features_unique_id.xlsx'
dest_filepath = '../data/01_raw/20221028_jb_features_type.csv'
```

```python
###load data###
ranges = pd.read_csv(source_file_ranges)
features = pd.read_excel(source_file_features)
features
```

```python
#set up data type lists
categoricals = ['persoonlijke_eigenschappen_uitstroom_verw_vlgs_klant', 'persoonlijke_eigenschappen_uitstroom_verw_vlgs_km',
               'persoonlijke_eigenschappen_taaleis_voldaan', 'persoonlijke_eigenschappen_spreektaal']
floats = ['adres_unieke_wijk_ratio']
bools = list(ranges.loc[(ranges['unique_values']==2), 'feature_name'])
```

```python
#set up defaults
features['type'] = 'numerical'
features['subtype'] = 'integer'
```

```python
#change type and subtype where appropriate
features['type'] =  np.where(features['feature_name'].isin(categoricals), 'categorical',features['type'])
features['subtype'] =  np.where(features['feature_name'].isin(categoricals), 'categorical',features['subtype'])
features['subtype'] =  np.where(features['feature_name'].isin(floats), 'float',features['subtype'])
features['type'] =  np.where(features['feature_name'].isin(bools), 'boolean',features['type'])
features['subtype'] =  np.where(features['feature_name'].isin(bools), 'boolean',features['subtype'])
features
```

```python
features.to_csv(dest_filepath, index=False)
```
